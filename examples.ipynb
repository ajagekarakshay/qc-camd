```python
from rdkit import Chem
```

```python
smiles = ['COc1c(Cl)nc(C)nc1NCC1CCN(c2cnccn2)CC1', 
        'O=C([C@@H](O)Cc1ccccc1)N1CCn2c(nnc2-c2ccccc2)C1', 
        'N#Cc1ccc(N[C@H]2CCCN(C(=O)c3ccccc3)C2)cc1', 
        'CC(NC(=O)CC1COC1)(c1ccccc1)c1ccccc1', 
        'COC[C@H](C)CC(=O)N1CCCC[C@@H]1CNC(=O)C(N)=O', 
        'O=C(c1cnc(O)[nH]1)N1CCO[C@]2(CCN(C(=O)C3CC3)C2)C1', 
        'CC(=O)c1cccc(C(=O)N2c3ccccc3OC[C@@H]2C)c1', 
        'Cc1ncn(C)c1C(=O)N1C[C@H](C)[C@@H](NC(=O)C(=O)NCC2CC2)C1', 
        'CCOC(=O)[C@H]1NC(=O)N(c2cc(-c3cccnc3)on2)C1=O', 
        'COc1cncc(C(=O)OC[C@H]2CCCN2CC(=O)O)n1', 
        'O=C(N[C@H]1[C@H]2C[C@H]1N(CCn1cncn1)C2)c1cc[nH]n1' ]

mols = [Chem.MolFromSmiles(smi) for smi in smiles]
```

```python
N = 3
Chem.Draw.MolsToGridImage(mols[:N], molsPerRow=3, subImgSize=(400,400))
```
